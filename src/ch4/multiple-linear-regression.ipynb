```python
import numpy as np
import matplotlib.pyplot as plt
from data import getData
```

```python
y, x1, x2 = getData()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
lr = 0.01
epochs = 2001
n = len(x1)
```

```python
a1 = 0
a2 = 0
b = 0
```

```python
for i in range(epochs):
    y_pred = a1 * x1 + a2 * x2 + b
    error = y - y_pred

    n = len(x1)
    a1_diff = 2 / n * sum(-x1 * error)
    a2_diff = 2 / n * sum(-x2 * error)
    b_diff = 2 / n * sum(-error)

    a1 -= lr * a1_diff
    a2 -= lr * a2_diff
    b -= lr * b_diff
    
    if i % 100 == 0:
        print("epoch=%.f, 1=%.04f, 2=%.04f =%.04f" % (i, a1, a2, b))

print(" : ", y)
print(" : ", y_pred)
```
