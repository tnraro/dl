```python
import numpy as np
import matplotlib.pyplot as plt
from data import getDataForKeras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
y, x = getDataForKeras()
```

```python
model = Sequential()
model.add(Dense(1, input_dim=2, activation="linear"))
model.compile(optimizer="sgd", loss="mse")
model.fit(x, y, epochs=2000)
```

```python
hour = 7
private_lesson = 4
prediction = model.predict([[hour, private_lesson]])
print("%.f시간 공부하고, %.f시간의 과외를 받은 경우의 예상 점수는 %.02f점입니다." % (hour, private_lesson, prediction))
```
